# آماده کردن دیتاست برای خوشه بندی 

```python
inertia_list=[]
for k in numpy.arange(1, 16):
    kmn= KMeans(n_clusters=k)
    kmn.fit(data.values)
    inertia_list.append(kmn.inertia_)
    
inertia_list
```

```python
# مشخص کردن تعداد خوشه های لازم
plt.plot(numpy.arange(1, 16), inertia_list, 'ro-')
plt.xlabel('number of clusters')
plt.ylabel('overlap')
plt.show()
```

```python
#بدون در نظر گرفتن تعداد خوشه های بهینه
#تعداد خوشه ها برابر با 15
# انجام عملیات خوشه بندی
kmn = KMeans(n_clusters=15)
kmn.fit(data.values)
labels = kmn.predict(data.values)
labels
```

```python
# برای یافتن 15 خوشه
#ترسیم نمودار براکندگی دو بعدی براساس مرگ روزانه و مبتلایان روزانه به تفکیک خوشه ها
xs = data.values[:,0]
ys = data.values[:,1]
plt.scatter(xs, ys, c=labels)
plt.legend(labels, loc='upper right')
plt.xlabel('first feature : A')
plt.ylabel('second feature : B')
plt.show()
```

```python
#تعداد خوشه ها برابر با 15
# ترسیم گراف پراکندگی سه بعدی
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
xs = data.values[:,0]
ys = data.values[:,1]
ax.scatter(xs, ys, c=labels)
ax.legend(labels, loc= 'upper right')
ax.set_xlabel('A')
ax.set_ylabel('B')
plt.show()
```
